```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(25,20,20)
```

```python
print (data.shape)
```

```python
plt.imshow(data[0], cmap=plt.cm.hot)
```

```python
 # Create figure with 3x3 sub-plots.
fig, axes = plt.subplots(5, 5)
#fig.subplots_adjust(hspace=0.3, wspace=0.3)
for i, ax in enumerate(axes.flat):
    # Plot image.
    ax.imshow(data[i], cmap=plt.cm.hot)
```
