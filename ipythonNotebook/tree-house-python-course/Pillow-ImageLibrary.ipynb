### Loading and showing images

```python
from PIL import Image
```

```python
lenna = Image.open('Lenna.png')
```

```python
lenna.show()
```

```python
print lenna.size
```

###Modify images

```python
#### Change modes(format) 
lenna.convert(mode='L').show() # 'L' Stands for Black and white mode
```

```python
### Resize
lenna.resize((100,100)).show()
```

```python
## Dhumb-nailing ( changes the image inplace rather than returning new)
lenna_thumb = lenna.copy()
lenna_thumb.thumbnail((100,100))
lenna_thumb
```

```python
## Cropping
box = (100,100,200,200)
```

```python
lenna.crop(box)
```

```python
lenna_thumb.rotate(90)
```

```python
## Copy paste image regions
region = lenna.crop(box).convert(mode='L')
lenna_2 = lenna.copy()
lenna_2.paste(region,box)
lenna_2.show()
```

```python
## Enhance, Filter images

from PIL import ImageEnhance, ImageFilter
lenna_enhance = ImageEnhance.Color(lenna)
lenna_enhance.enhance(0.5).show()
```

```python
lenna.filter(ImageFilter.GaussianBlur(radius=10)).show()
    
```

```python
lenna.filter(ImageFilter.EMBOSS).show()
```
