```python
import graphlab
```

```python
dataSet = graphlab.SFrame('amazon_baby.gl/')
```

```python
dataSet.show()
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
dataSet['word_count'] = graphlab.text_analytics.count_words(dataSet['review'])
```

```python
dataSet.head()
```

```python
dataSet['awesome'] = dataSet['word_count'].apply(lambda wordCountDict: wordCountDict['awesome'] if 'awesome' in wordCountDict else 0 )
```

```python
dataSet[dataSet['awesome']>3]['review']
```

```python
for eachSelectWord in selected_words:
    dataSet[eachSelectWord] = dataSet['word_count'].apply(lambda wordCountDict: wordCountDict[eachSelectWord] if eachSelectWord in wordCountDict else 0 )
```

```python
dataSet.head()
```

```python
sumSelectWords = {}
for eachSelectWord in selected_words:
    sumSelectWords[eachSelectWord] = dataSet[eachSelectWord].sum()
```

```python
print sumSelectWords
```

```python
dataSet = dataSet[dataSet['rating'] != 3]
dataSet['sentiment'] = dataSet['rating'] >=4
train_data,test_data = dataSet.random_split(.8, seed=0)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=selected_words,
                                                     validation_set=test_data)
```

```python
selected_words_model['coefficients']
```

```python
selected_words_model['coefficients'].sort('value',ascending=False)
```

```python
selected_words_model.evaluate(test_data)
```

```python
diaper_champ_reviews = dataSet[dataSet['name'] == 'Baby Trend Diaper Champ']
```

```python
diaper_champ_reviews.show()
```

```python
diaper_champ_reviews['predicted_sentiment'] = selected_words_model.predict(diaper_champ_reviews, output_type='probability')
```
